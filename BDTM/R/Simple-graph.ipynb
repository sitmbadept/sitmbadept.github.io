```r
data <- read.csv("marketing-spend.csv")
```

```r
str(data)
```

```r
# Show Histogram for Sales
hist(data$Sales,
     col='green',
     main="Sales Distribution")
```

```r
# Show Boxplot 
boxplot(data$Sales)
```

```r
# Display Pie chart for Monthly Spend
pie(data$Spend, data$Month)
```

```r
# Display Bar graph for Monthly Spend
barplot(data$Spend,
        names.arg = data$Month)
```

```r
# Display Bar graph for Monthly Spend
barplot(data$Spend,
        names.arg = data$Month,
        col=rainbow(12))
```

```r
# Display Bar graph for Monthly Spend
barplot(data$Spend,
        names.arg = data$Month,
        col=rainbow(12),
        horiz=TRUE)
```

```r
# Display Scatter plot Spend vs Sales
plot(data$Spend,
     data$Sales,
     col="red")
```

```r
# Display Line Chart Month vs Spend
plot(data$Month,
     data$Spend,
     col='red',
     type='b') # Type : p l b c o h s S n
```
